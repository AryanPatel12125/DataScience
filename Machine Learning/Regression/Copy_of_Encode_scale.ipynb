```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler
```

```python
# df = pd.read_excel('/content/File1 (copy).xlsx')
```

```python
d = {"Gender" : ["M", "F", "F", "F", "M"],
     "Subjects" : ["Hindi", "Math", "Science", "Science", "Math"],
     "Marks" : [100, 65, 33, 43, 55],
     "Result" : ["P", "P", "F", "P", "P"]
     }

df = pd.DataFrame.from_dict(d)
```

```python
df.head()
```

```python
encoder = OneHotEncoder(sparse = False, handle_unknown = 'ignore')
encoder.fit(df)
df1 = pd.DataFrame(encoder.transform(df), columns = encoder.get_feature_names_out())
```
