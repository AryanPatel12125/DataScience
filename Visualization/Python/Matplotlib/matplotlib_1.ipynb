**Scatterplot**

```python
plt.scatter(x = tips['total_bill'], y = tips['tip'])

plt.xlabel('total bill')
plt.ylabel('tips')
plt.show()
```

```python
plt.scatter(x = tips['total_bill'], y = tips['tip'], c = 'green')

plt.xlabel('total bill')
plt.ylabel('tips')
plt.show()
```

```python
t = plt.scatter(x = tips['total_bill'], y = tips['tip'], c = 'green')
s = plt.scatter(x = tips['total_bill'], y = tips['size'], c = 'red')

plt.xlabel('total bill')
plt.ylabel('tips')
plt.show()
```

```python
t = plt.scatter(x = tips['total_bill'], y = tips['tip'], c = 'green')
s = plt.scatter(x = tips['total_bill'], y = tips['size'], c = 'red')

plt.legend((t, s), ('tips', 'size'))
plt.xlabel('total bill')
plt.ylabel('tips')
plt.show()
```

```python
plt.plot(tips['total_bill'], tips['tip'])
plt.xlabel('total bill')
plt.ylabel('tip')
plt.show()
```

```python
plt.plot(np.arange(0, 1, 0.2), (np.arange(0, 1, 0.2)**2))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(tips['tip'])
plt.xlabel('tip received')
plt.show()
```

```python
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(tips['tip'], bins = 20)
plt.xlabel('tip received')
plt.show()
```

```python
plt.boxplot([tips['size'], tips['tip'], tips['total_bill']])
plt.xlabel('size')
plt.show()
```

```python
plt.bar(tips['sex'], tips['tip'], color = 'green')
plt.xlabel('sex')
plt.ylabel('tip')
plt.show()
```

```python
plt.bar(tips['sex'], tips['tip'], color = 'green', width = 0.4)
plt.xlabel('sex')
plt.ylabel('tip')
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([2,4,6,8])

plt.bar(x,y)
plt.show()
```

# Pia chart

```python
#labels
y = np.array([20,35,10,45])
mylabels = ["USA", "India", "Russia", "China"]

plt.pie(y, labels=mylabels,autopct='%1.2f%%')  # autopct to show percentage
plt.show()
```

```python
#legend
plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

```python
#legend with title
plt.pie(y, labels = mylabels)
plt.legend(title = "Country")
plt.show()
```
