RelPlot is used when we want to observe the relationship between 2 features i.e. multivariate analysis. Scatterplot is the default kind in relplot.

When using Seaborn we need to explicitly specify the parameters on the x and y axis.

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips)
```

```python
sns.relplot(x="total_bill", y="tip", kind="line", data=tips)
```

**Hue** 

Grouping variable that will produce elements with different colors. Can be either categorical or numeric, although color mapping will behave differently in latter case.

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

**Style** 

Grouping variable that will produce elements with different styles. Can have a numeric dtype but will always be treated as categorical.

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips)
```

**Kind**

Kind of plot to draw, corresponding to a seaborn relational plot. Options are "scatter" or "line".

```python
sns.relplot(x="total_bill", y="tip", kind="line", data=tips)
```

**Plots for Univariate analysis**

Univariate analysis is the simplest form of analyzing data. “Uni” means “one”, so in other words your data has only one variable. It doesn't deal with causes or relationships and it's major purpose is to describe; It takes data, summarizes that data and finds patterns in the data.

Histogram and bar plots can be used for univaraite analysis.

Distplot is deprecated so better to use histplot.

Histplot is used to get the distribution of the feature.

**stat**

statistic to calculate for each bin

```python
sns.histplot(x = tips['tip'], stat = 'count')
```

```python
sns.histplot(x = tips['tip'], stat = 'percent')
```

**binwidth**

We can also change the bin width for visual purposes.

```python
sns.histplot(x = tips['tip'], stat = 'percent', binwidth = 0.6)
```

**kde**

Kernel Density Estimate to get a trend of the distribution

```python
sns.histplot(x = tips['tip'], stat = 'percent', kde = True)
```

**Countplot**

A count plot can be thought of as a histogram across a categorical, instead of numerical, variable.

```python
sns.countplot(x ='smoker',data=tips)
```

**Hue**

We can also include hue to add more richness and insights into one graph.

```python
sns.countplot(x = 'smoker', hue = 'sex' , data = tips)
```

Plot features on y axis if x axis is getting crowded.

```python
sns.countplot(y = 'day', hue = 'smoker' , data = tips)
```

Seaborn's jointplot displays a relationship between 2 variables (bivariate) as well as 1D profiles (univariate) in the margins.

By default, Joint Plot uses Scatter Plot and Histogram. Joint Plot can also display data using Kernel Density Estimate (KDE)

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = tips, kind = 'reg')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = tips, hue = 'sex')
```

**Boxplot**

Used when we want to obtain distribution as well as various quantiles.

Assign a single feature for univariate analysis and multiple features for multivariate analysis

```python
sns.boxplot(x ='tip', data=tips)
```

```python
sns.boxplot(x ='sex', y = 'total_bill', data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
sns.boxplot(x="total_bill", y="day", hue="sex",data=tips)
```

A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. 

Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(x="total_bill", y="sex", data=tips,palette='rainbow')
```
